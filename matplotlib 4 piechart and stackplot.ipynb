```python
from matplotlib import pyplot as plt
plt.style.use("fivethirtyeight")
import matplotlib as mpl
mpl.style.use("default")
```

```python
slices = [120, 80, 30, 20]

labels = ["Sixty", "Forty", "Extra", "Minuse"]
colors = ["#008fd5", "#fc4f30", "#e5ae37", "#6d904f"]

#wedgeprops apvadus padaro schemai
plt.pie(slices, labels=labels, colors = colors, wedgeprops = {"edgecolor":"black"})


plt.title("pie chart")
plt.tight_layout()
plt.show()

# Colors:
#     Blue = #008fd5
#     Red = #fc4f30
#     Yellow = #e5ae37
#     Green = #6d904f
```

```python
# atstumti viena is duomenu, uzdeti seseli, pasukti diagrama, uzdeti reiksmes
slices = [59219, 55466, 47544, 36443, 35917]
labels = ['JavaScript', 'HTML/CSS', 'SQL', 'Python', 'Java' ]
explode = [0, 0, 0, 0.1, 0]

plt.pie(slices, labels=labels, explode = explode, shadow= True, startangle = 90, autopct = "%1.1f%%", wedgeprops = {"edgecolor":"black"})


plt.title("pie chart")
plt.tight_layout()
plt.show()
```

```python
#stackplots kiek kiekvienas ideda indelio tam tikru laiko momentu
import matplotlib as mpl
mpl.style.use("default")
minutes = [1, 2 , 3 , 4  ,5 , 6 , 7 , 8, 9]
player1 = [8, 6, 5, 5, 4, 2, 1, 1, 0]
player2 = [0, 1, 2, 2, 2, 4, 4, 4, 4]
player3 = [0, 1, 1, 1, 2, 2, 3, 3, 4]

labels = ['player1', 'player2','player3']
colors = ["#008fd5", "#fc4f30", "#6d904f"]

plt.stackplot(minutes, player1, player2, player3, colors= colors,  labels = labels )

# 7 procentus nuo apacios ir 5 procentus nuo kaires arba galima rasyti 
# su string upper left arba lower left
plt.legend(loc=(0.07, 0.05))

plt.title("zaidejai")
plt.tight_layout()
```
