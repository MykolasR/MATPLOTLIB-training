```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
data = pd.read_csv("C:/Users/Vartotojas/Desktop/kursai online/Matplotlib/data2.csv")

data.head(5)
```

```python

ages = data["Age"]
dev_salaries = data["All_Devs"]
py_salaries = data["Python"]
js_salaries = data["JavaScript"]

plt.plot(ages, dev_salaries, color = "#444444", 
         linestyle = "--", label = "All Devs")

plt.plot(ages, py_salaries, label = "Python")


# uzpildyti tarp kreiviu, alpha nustato ryskuma, conditionals su where ir interpolate gali padalinti spalvomis ir pan.
overall_median = 57287

plt.fill_between(ages, py_salaries, overall_median, 
                 where= (py_salaries> overall_median),
                 interpolate=True, alpha=0.25, color= "red")

plt.fill_between(ages, py_salaries, overall_median, 
                 where= (py_salaries < overall_median),
                 interpolate=True, alpha=0.25)

plt.legend()

plt.title("kazkokia lentele")
plt.xlabel("Ages")
plt.ylabel("median salary")
```
