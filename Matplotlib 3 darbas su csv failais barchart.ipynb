```python
import csv
import numpy as np
from collections import Counter
from matplotlib import pyplot as plt
plt.style.use("fivethirtyeight")
import matplotlib as mpl
mpl.style.use("default")
import pandas as pd
```

```python
#variantas NO1 su import csv
#ikeliame faila
with open ("C:/Users/Vartotojas/Desktop/kursai online/data.csv") as csv_file:
    csv_reader = csv.DictReader(csv_file)
# sukuriame counteri  
    lang_counter = Counter()
# idedame is failo skilti languages worked with, kad counteris skaiciuotu    
    for row in csv_reader:
        lang_counter.update(row['LanguagesWorkedWith'].split(";"))
# suformuojame du listus languages ir popularity, kad veliau padarytume plot
languages = []        
popularity = []
        
for item  in lang_counter.most_common(15):
    languages.append(item[0])
    popularity.append(item[1])
    
print(languages)
print(popularity)

#apverciame listus

languages.reverse()
popularity.reverse()

#padarome schema su barh metodu
plt.barh(languages, popularity)
plt.title("kalbos pop")
plt.ylabel("lang")
plt.xlabel("pop")
plt.grid(linewidth=1, color="red")
```
